```python
df = pd.read_csv('data/nyu_capstone_data/tripadvisor_100k_poi_reviews.csv')
df = df.drop('Unnamed: 0', axis = 1)
df
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
# fp = open('nyu_capstone_data/tripadvisor_100k_poi_reviews.csv')
# data = fp.read()
data = df['reviewtext']
# print ('\n-----\n'.join(tokenizer.tokenize(data)))

tokenized_data = data.apply(tokenizer.tokenize)
tokenized_data
```

```python
tokenized_data[0]
```

## Break down list of sentences into sentences and duplicate the features (Unpivoting)

```python
from tqdm import tqdm
```

```python
index_list = []
sentences = []

for index, value in tqdm(tokenized_data.items()):
    for sent in value:
        index_list.append(index)
        sentences.append(sent)
    
```

```python
temp = pd.DataFrame({'sentence': sentences, 'original_index': index_list})
```

```python
combinedDF = temp.merge(df, left_on='original_index', right_index=True)
```

```python
combinedDF.drop(columns=['reviewtext', 'Sentences', 'original_index'], inplace=True)
```
